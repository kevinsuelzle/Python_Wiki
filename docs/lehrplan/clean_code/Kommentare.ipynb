```python
class DatenAnalysator:
    """
    Eine Klasse zur Durchf√ºhrung einfacher statistischer Analysen auf einer Liste von numerischen Daten.

    Attributes
    ----------
    daten : list of int/float
        Liste der numerischen Daten, mit denen die Analyse durchgef√ºhrt wird.

    Methods
    -------
    mittelwert()
        Berechnet den Mittelwert der Daten.
    maximum()
        Findet den maximalen Wert in den Daten.
    minimum()
        Findet den minimalen Wert in den Daten.
    """

    def __init__(self, daten):
        """
        Initialisiert den DatenAnalysator mit einer Liste von Daten.

        Parameters
        ----------
        daten : list of int/float
            Die zu analysierenden numerischen Daten.
        """
        self.daten = daten

    def mittelwert(self):
        """
        Berechnet den Mittelwert der Daten.

        Returns
        -------
        float
            Der Mittelwert der Daten.
        """
        return sum(self.daten) / len(self.daten)

    def maximum(self):
        """
        Findet den maximalen Wert in den Daten.

        Returns
        -------
        int/float
            Der maximale Wert in den Daten.
        """
        return max(self.daten)

    def minimum(self):
        """
        Findet den minimalen Wert in den Daten.

        Returns
        -------
        int/float
            Der minimale Wert in den Daten.
        """
        return min(self.daten)

def daten_normalisieren(daten):
    """
    Normalisiert eine Liste von numerischen Daten, so dass alle Werte zwischen 0 und 1 liegen.

    Parameters
    ----------
    daten : list of int/float
        Die Liste der zu normalisierenden numerischen Daten.

    Returns
    -------
    list of float
        Eine Liste der normalisierten Daten, wobei jeder Wert zwischen 0 und 1 liegt.
    """
    max_wert = max(daten)
    min_wert = min(daten)
    return [(x - min_wert) / (max_wert - min_wert) for x in daten]
```
